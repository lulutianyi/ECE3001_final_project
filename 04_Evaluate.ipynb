# 04 — Evaluate

Evaluate best checkpoint on test split.

```python
import matplotlib.pyplot as plt
import numpy as np
from evaluate import evaluate_on_split

DATASET_ROOT = "./data/ravdess"  # TODO(student)
SPLIT_JSON = "./splits/split_seed42.json"
CKPT_PATH = "./outputs/exp1/best.pt"

out = evaluate_on_split(DATASET_ROOT, SPLIT_JSON, "test", CKPT_PATH, device="cuda")
print("Test accuracy:", out["acc"])
print(out["report"])
cm = out["confusion_matrix"]
```

```python
plt.figure(figsize=(8,8))
plt.imshow(cm.astype(np.float32), aspect="auto", origin="upper")
plt.title("Confusion Matrix (Test)")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.colorbar()
plt.show()
```
